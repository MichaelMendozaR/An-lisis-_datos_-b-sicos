```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
productos = ['Pr A', 'Pr B', 'Pr C','Pr D','Pr E','Pr F','Pr G','Pr H',]
Bogota = [150, 100, 90,234,321,455,222,111] # Changed to 3 elements to match productos
Cali = [200, 180, 150,222,333,444,555,666] # Changed to 3 elements to match productos
Bucaramanga = [250, 220, 300,400,432,234,123,234] # Changed to 3 elements to match productos

plt.bar(productos, Bogota, label='Bogota', color='blue')
plt.bar(productos, Cali, bottom= Bogota, label='Cali', color='green')
plt.bar(productos, Bucaramanga, bottom=[i+j for i,j in zip(Bogota, Cali)], label='Bucaramanga',color='red')

plt.xlabel('Productos')
plt.ylabel('Ventas (Unidades)')
plt.title('Comparaci√≥n de Ventas por Producto y Ciudad')
plt.legend()
```

```python
datos = pd.DataFrame({
'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo','Junio', 'Julio', 'Agosto'],
'Papa (Cali)': [100, 150, 200, 250, 300,400,300,150],
'Cebolla(Bogota)': [90, 120, 160, 200, 240,100,200,150],
'Tomate(Cartagena)': [80, 110, 150, 190, 230,330,350,240]
})

sns.set(style="whitegrid")

plt.plot(datos['Mes'], datos['Papa (Cali)'], marker='o', color='blue', label='Papa (Cali)')
plt.plot(datos['Mes'], datos['Cebolla(Bogota)'], marker='s', color='green',
label='Cebolla(Bogota)')
plt.plot(datos['Mes'], datos['Tomate(Cartagena)'], marker='^', color='red',
label='Tomate(Cartagena)')

plt.title('Evoluci√≥n de ventas por producto y ciudad', fontsize=14)
plt.xlabel('Mes', fontsize=12)
plt.ylabel('Ventas (Unidades)', fontsize=12)

plt.legend()

plt.show()
```

```python
datos = pd.DataFrame({
'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre',
'Octubre', 'Noviembre', 'Diciembre'],
'SUV': [120, 130, 135, 140, 150, 160, 170, 180, 175, 185, 190, 200],
'Sedan': [100, 90, 95, 110, 120, 130, 140, 145, 150, 160, 170, 180],
'Hatchback': [80, 85, 90, 95, 100, 105, 110, 115, 120, 125, 130, 135]
})

sns.set(style="whitegrid")

plt.plot(datos['Mes'], datos['SUV'], label='SUV', marker='o', color='blue', linestyle='--')
plt.plot(datos['Mes'], datos['Sedan'], label='Sedan', marker='s', color='green', linestyle='-.')
plt.plot(datos['Mes'], datos['Hatchback'], label='Hatchback', marker='^', color='red',
linestyle=':')

plt.title('Ventas Mensuales de Veh√≠culos por Tipo', fontsize=14)
plt.xlabel('Meses', fontsize=12)
plt.ylabel('Ventas (unidades)', fontsize=12)

plt.xticks(rotation=45, ha='right')

plt.legend()
plt.tight_layout()
plt.show()
```
